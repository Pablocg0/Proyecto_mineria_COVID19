```python
import string
with open('./tweetsTXT/tweets0.txt') as f:
    lines = f.read()#.splitlines()


salida = re.sub( '[% s]' % re.escape(string.punctuation), '', lines)
```

```python
salida1=re.findall('\w+',salida)
salida1
```

```python
len(salida1)
```

```python
#Palabras distintas
len(set(salida1))
```

```python
#lematizamos las palabras
import spacy 
nlp=spacy.load('es_core_news_sm')
```

```python
salida2=[]
for i in salida1:
    doc= nlp(i)
    for token in doc:
        #print(token.text, token.lemma_)
        salida2.append(token.lemma_)
```

```python
with open('./tweetsTXT/tweets0.txt') as f:
    lines2 = f.read().splitlines()
```

```python
 
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
from sklearn.metrics import adjusted_rand_score
```

```python
from nltk.corpus import stopwords
stopwords = [word for word in stopwords.words('spanish')]

vectorizer = TfidfVectorizer(stopwords)
X = vectorizer.fit_transform(lines2)
true_k = 4
model = KMeans(n_clusters=true_k, init='k-means++', max_iter=100, n_init=1)
model.fit(X)
```

```python
print("Top terms per cluster:")
order_centroids = model.cluster_centers_.argsort()[:, ::-1]
terms = vectorizer.get_feature_names()
for i in range(true_k):
    print("Cluster %d:" % i, 'elementos= ',len(order_centroids[i])),
    for ind in order_centroids[i, :10]:
        print(' %s' % terms[ind]),
```
