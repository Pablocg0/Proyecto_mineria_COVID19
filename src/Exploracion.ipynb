```python
import re

def deEmojify(inputString):
    return inputString.encode('ascii', 'ignore').decode('ascii')


def dataImport(filename):
    info = []
    for xs in filename:
        data = open(xs)
        for r in data:
            r = r.replace('\n', '')
            r = re.sub('https\W*t.co/\w*','',r)
            info.append(r.encode('ascii', 'ignore').decode('ascii'))
    return info

data = dataImport(["../data/tweets0.txt","../data/tweets1.txt","../data/tweets2.txt","../data/tweets3.txt","../data/tweets4.txt","../data/tweets5.txt"])
print(len(data))
```

```python
from sklearn.feature_extraction.text import CountVectorizer
count_vect = CountVectorizer() 
X_train_counts = count_vect.fit_transform(data) 
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt


y_pred = KMeans(n_clusters=4 ).fit_predict(X_train_counts)


print(y_pred)
```

```python
x_pred =  range(0, len(y_pred))


plt.figure(figsize=(22.2, 11.4))
plt.scatter(x = x_pred, y = y_pred)
#plt.plot(y_pred , 'o')
plt.show
```

```python
k0 = []
k1 = []
k2 = []
k3 = []

for xs in x_pred:
    if y_pred[xs] == 0:
        k0.append(data[xs])
    elif y_pred[xs] == 1:
        k1.append(data[xs])
    elif y_pred[xs] == 2:
        k2.append(data[xs])
    elif y_pred[xs] == 3:
        k3.append(data[xs])
        
        
print(len(k0))
print(len(k1))
print(len(k2))
print(len(k3))
        
    
```

```python
def escribe(file, data):
    file1 = open(file,"w") 
    for xs in data:
        file1.write(xs+'\n')
    file1.close()
    
escribe('k0.txt',k0)
escribe('k1.txt',k1)
escribe('k2.txt',k2)
escribe('k3.txt',k3)
```

```python
xk31 = count_vect.transform(k0)
y_pred1 = KMeans(n_clusters=4 ).fit_predict(xk31)


print(y_pred1)
```

```python
x_pred =  range(0, len(y_pred1))


plt.figure(figsize=(22.2, 11.4))
plt.scatter(x = x_pred, y = y_pred1)
#plt.plot(y_pred , 'o')
plt.show
```
